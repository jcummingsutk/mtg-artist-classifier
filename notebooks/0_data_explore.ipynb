```python
from torchvision.datasets import ImageFolder
from torch.utils.data import Dataset
import matplotlib.pyplot as plt
import os
from PIL import Image
import seaborn as sns
import pandas as pd

sns.set_style("darkgrid")

train_dataset = ImageFolder(
    root=os.path.join(
        "data",
        "model_training_and_eval",
        "train_images"
    )
)
eval_dataset = ImageFolder(
    root=os.path.join(
        "data",
        "model_training_and_eval",
        "eval_images"
    )
)
test_dataset = ImageFolder(
    root=os.path.join(
        "data",
        "model_training_and_eval",
        "test_images"
    )
)
```

```python
eval_dataset[0]
```

```python
total_images = train_dataset + eval_dataset + test_dataset
```

```python
test_dataset.classes
```

```python
total_images
```

```python
columns = ["h_pixels", "v_pixels", "class"]
data = [
    [img[0].size[0], img[0].size[1], test_dataset.classes[img[1]]]
    for img in total_images
]
image_info_df = pd.DataFrame(data, columns=columns)
```

```python
sns.countplot(image_info_df, x="h_pixels", hue="class")
```

```python
sns.countplot(image_info_df, x="v_pixels", hue="class")
```
